2). Get rid of the columns that have more than 97% missing values.

```python
id_to_drop = set(df.columns[df.isnull().mean() > 0.97])
df = df.drop( id_to_drop , axis = 1)
df.head()
```

3). Get rid of the unnecessary columns. HINT: Those columns with "time" in the name and those that are obviously unnecessary.

```python
#drop time cols 
cols = df.columns
for i in cols : 
    if 'time' in i : df = df.drop( i , axis = 1)
df.head()
```

```python
#drop unnecessary 
df = df.drop(['Unnamed: 0','user_name','new_window'],axis =1 )
```

4). Label encode the responde variable. HINT: use preprocessing.LabelEncoder(). 

```python
#check if all the data is numerical
df.select_dtypes(['number'])
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['classe'] = encoder.fit_transform(df['classe'])
df.head()
```

5). Carry out min-max scaling of the exploratory variables. HINT: use preprocessing.MinMaxScaler().

```python
# <Your code goes in here>
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
for col in df.columns : 
    x = df[col].values.reshape(-1, 1) #returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    df[col] = x_scaled
  
```

```python
#importing relevant libraries
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
#features extraction
x=df[["pitch_forearm", "magnet_arm_x", "accel_arm_x", "total_accel_forearm", "magnet_dumbbell_z","accel_dumbbell_x","roll_arm","pitch_dumbbell","total_accel_belt","accel_dumbbell_z","magnet_dumbbell_x","roll_belt","yaw_arm","accel_arm_z"]]
y= (df["classe"]).astype(int)

#splitting data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)

#applying tree algorithm
tree = tree.DecisionTreeClassifier()  
tree.fit(x_train, y_train)   #fitting our model
y_pred=tree.predict(x_test)   # evaluating our model
print("score:{}".format(accuracy_score(y_test, y_pred)))
```
